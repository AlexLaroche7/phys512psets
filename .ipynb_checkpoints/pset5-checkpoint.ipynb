```python
import numpy as np
from matplotlib import pyplot as plt
```

### Problem 1

```python
def shift_array(arr,shift):
    '''
    Returns array with 'arr' elements shifted to the right by 'shift'
    
    :param arr: numpy.ndarray to shift
    :param shift: int by which to shift 'arr'
    '''
    if shift > len(arr): 
        shift%=len(arr) #modulo array length so that shift is periodic if shift > array length
    ker=np.zeros(len(arr)) 
    ker[shift]=1 #set shift index in kernel to 1
    return np.fft.irfft(np.fft.fft(arr)*np.fft.fft(ker),len(arr)) #convolution shifts array by shift

def gauss(x,A=1,mu=0,sig=1):
    return A*np.exp(-0.5*((x-mu)/sig)**2)

x=np.linspace(-4,4,1001)
f=gauss(x)
plt.plot(x,f,label='Gaussian')

shift=int(len(x)/2) #half of array length
f_shift=shift_array(f,shift) #shift the gaussian
plt.plot(x,f_shift,label='shifted Gaussian')

plt.legend()
plt.savefig('gauss_shift.png')
plt.show()
```

### Problem 2

```python
def corr_func(f,g):
    '''
    Returns correlation function of f and g
    
    :param f,g: numpy.ndarrays to correlate
    '''
    corr=np.fft.irfft(np.fft.fft(f)*np.conj(np.fft.fft(g)),len(f)) 
    return corr/np.max(corr) #normalize

corr=corr_func(f,f)
plt.plot(x,f,label='Gaussian')
plt.plot(x,corr,label='Correlation')
plt.legend()
plt.show()
```

### Problem 3

```python
def shift_corr(f,shift):
    return corr_func(f,shift_array(f,shift))

plt.plot(x,corr,label='No shift')

shifts=[200,400,600,800]
for shift in shifts:
    label='shift by '+str(shift)
    plt.plot(x,shift_corr(f,shift),label=label)
    
plt.legend()
plt.show()
```

The correlation function is shifted by the shift of the second Gaussian, which makes sense. The amplitude of the correlation function is shift independent. This is not surprising because the DFT is circulant.

### Problem 4

```python
def conv_safe(f,g,pad):
    if len(f)<len(g):
        shift=len(g)-len(f)
        f=np.append(f,np.zeros(shift))
    if len(f)>len(g):
        shift=len(f)-len(g)
        g=np.append(g,np.zeros(shift))
    zeros=np.zeros(pad)
    f0=np.append(f,zeros)
    g0=np.append(g,zeros)
    conv=np.fft.irfft(np.fft.rfft(f0)*np.fft.rfft(g0))
    return conv

#make gaussians of different lengths
x1=np.linspace(-4,4,1001) 
x2=np.linspace(-4,4,2001)
f=gauss(x1)
g=gauss(x2)

#convolve the gaussians safely by padding with zeros
pad=1001
conv=conv_safe(f,g,pad)
plt.plot(conv)
plt.show()

print('The length of the convolution is',len(conv))
print('max length of (f,g) + length pad =',max(len(f),len(g))+pad)
```

The length of the safe convolution is the length of the longest function array plus the length of the zero padding. As expected, the convolution of a gaussian is a gaussian and it is centered about the midpoint since the inputted gaussians are centered. 

### Problem 5

(a) We recognize that 

$$ \sum_{x=0}^{N-1}\exp(-2\pi i kx/N) $$

is a geometric series. Since

$$ \sum_{x=0}^{N-1} r^k = \frac{1-r^{N}}{1-r}, $$

for $r=\exp(-2\pi ik/N)$ we have

$$ \sum_{x=0}^{N-1}\exp(-2\pi i kx/N) = \frac{1-\exp(-2\pi ik)}{1-\exp(-2\pi ik/N)}. $$

(b) In the limit that $k\to0,$ 

$$ \sum_{x=0}^{N-1}\exp(-2\pi i kx/N) \to \lim_{k\to0} \frac{1-\exp(-2\pi ik)}{1-\exp(-2\pi ik/N)} = N\lim_{k\to0} \frac{\exp(-2\pi ik)}{\exp(-2\pi ik/N)} = N\lim_{k\to0}\exp(-2\pi ik(1+1/N)) = N,$$

where we make us of L'Hopitals Rule in the first equality. If $k\%N\neq0,$ $\exp(-2\pi ik)=1,$ so the numerator is zero, and $\exp(-2\pi ik/N)\neq0$ since $k/N$ is not an integer, so the denominator is non-zero. Hence, 

$$ \sum_{x=0}^{N-1}\exp(-2\pi i kx/N) = 0. $$

(c) The expression we just derived is equivalent to the DFT of a sine wave of wave number $k.$ We will now show that for non-integer $k,$ the DFT behaves as expected.

```python
def dft_estimate(f):
    N=len(f)
    x=np.arange(N)
    dft=np.zeros(N,dtype=np.complex128)
    for k in range(N):
        for i in range(N):
            dft[k]+=f[i]*np.exp(-2*np.pi*1j*k*x[i]/N)
    return dft

N=100
x=np.arange(N)
k=22.89
f=np.sin(2*np.pi*k*x/N)
dft_est=dft_estimate(f)
```

Let's plot the DFT using np.fft and compare it to our DFT estimate.

```python
dft=np.fft.rfft(f)
k=np.arange(len(f))[:N//2+1]

plt.plot(k,np.abs(dft_est[:N//2+1]),label='Estimate')
plt.plot(k,np.abs(dft),'--',label='np.fft')
plt.legend()
plt.show()

plt.plot(k,np.real(dft_est[:N//2+1]-dft),label='DFT error')
plt.legend()
plt.show()
```

The estimate agrees with np.fft to $O(10^{-13}),$ which is pretty good. The DFTs don't look like delta functions because of spectral leakage. We do still see a sharp peak at the wavenumber of the sine function.

(e) We will now convolve our function with a window function before taking the DFT to see if we can minimize spectral leakage.

```python
window=0.5-0.5*np.cos(2*np.pi*x/N)
f_window=f*window
dft_est_window=dft_estimate(f_window)
```

```python
dft_window=np.fft.rfft(f_window)
k=np.arange(len(f))[:N//2+1]

plt.plot(k,np.abs(dft_est_window[:N//2+1]),label='Estimate')
plt.plot(k,np.abs(dft_window),'--',label='np.fft')
plt.plot(k,np.abs(dft_est[:N//2+1]),label='Estimate window')
plt.plot(k,np.abs(dft),'--',label='np.fft window')
plt.legend()
plt.show()
```

We observe that applying the window function makes both the estimated and np.fft DFTs more like delta functions because their peaks are sharper. We have succesfully decreased spectral leakage.

(e) We will now look at the FT of the window function

```python
ftw=np.fft.fft(window)
arr=np.zeros(N)
arr[0]=N/2
arr[1]=-N/4
arr[-1]=-N/4
plt.plot(ftw,label='np.fft')
plt.plot(arr,'--',label='correct')
plt.legend()
plt.show()
```

The FT of the window function behaves as expected. We may apply the window function by 

$$ F[k] = \frac{F[k-1]}{4}+\frac{F[k]}{2}+\frac{F[k+1]}{4},$$

in Fourier space. Let's check this.

```python
fft=np.fft.rfft(f)
fft_win_add=fft/2+np.roll(fft,1)/4+np.roll(fft,-1)/4

plt.plot(k,np.abs(fft_win_add),label='add neighbors')
plt.plot(k,np.abs(np.fft.rfft(f*window)),label=r'f$\times$window')
plt.legend()
plt.show()
```

We observe that applying the window function through the above expression in Fourier space is equivalent to multiplying the data by the window function in real space.

(b) Let's generate a random walk and check the power spectra behaviour. 

```python
n=int(1e4)
num_walks=1000
walk=np.cumsum(np.random.randn(n))
power=np.abs(np.fft.rfft(walk))**2 
power=power[1:] #ignore k=0
plt.loglog(power[:20],label='Walk power')
k=(np.arange(n)[1:])*1.0
k2=k**-2
plt.loglog(power[0]*k2[:20],'r--',label=r'$k^{-2}$')
plt.legend()
plt.savefig('rw_ps.png')
plt.show()
```

As expected, the walk power roughly follows a $k^{-2}$ power law.
